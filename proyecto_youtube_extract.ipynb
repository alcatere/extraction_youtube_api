```python
def informacion_especifica(Video_ID, videos_database):
    url_video = f'https://www.googleapis.com/youtube/v3/videos?key={API_KEY}&part=snippet,contentDetails,statistics&id={Video_ID}'
    response_video = requests.get(url_video).json()
    Duration = response_video['items'][0]['contentDetails']['duration']
    Likes = response_video['items'][0]['statistics']['likeCount']
    Views = response_video['items'][0]['statistics']['viewCount']
    # DisLikes = response_video['items'][0]['statistics']['dislikeCount'] NO SE PUEDE AGREGAR A MENOS DE QUE SEAS EL PROPIETARIO DEL CANAL
    Comments_Count = response_video['items'][0]['statistics']['commentCount']
    Favorites_Count = response_video['items'][0]['statistics']['favoriteCount']
    try:
        Tags = ', '.join(response_video['items'][0]['snippet']['tags'])
    except KeyError:
        print(f'El video {Video_ID} no tiene tags')
        Tags = np.nan

    # Se agregan al diccionario de videos anterior
    videos_database['Duration'] = Duration
    videos_database['Views'] = Views
    videos_database['Likes'] = Likes
    videos_database['Comments_Count'] = Comments_Count
    videos_database['Favorites_Count'] = Favorites_Count
    videos_database['Tags'] = Tags
    
    return videos_database
```

```python
# Obtenemos los 50 videos que han tenido una mayor relevancia
def obtener_informacion_videos_relevance(API_KEY, CHANNEL_ID):
    df_videos = pd.DataFrame(columns=['Video_ID','CHANNEL_ID','Extraction_Date','Title','Published_At','Description','Thumbnail_url','Duration','Likes','Views','Comments_Count','Favorites_Count','Tags'])
    url = f'https://www.googleapis.com/youtube/v3/search?key={API_KEY}&channelId={CHANNEL_ID}&part=snippet,id&maxResults=50' # Para buscar videos
    response = requests.get(url).json()
    for video in response['items']:
        time.sleep(1)
        if video['id']['kind'] == 'youtube#video':
            Video_ID = video['id']['videoId'] # ID del video
            Extraction_Date = datetime.date.today() # Dia y hora de la extraccion
            Title = video['snippet']['title'] # Titulo del video
            Published_At = video['snippet']['publishedAt'] # Fecha de publicacion
            Description = video['snippet']['description'] # Descripcion del video
            Thumbnail_url = video['snippet']['thumbnails']['default']['url'] # El url del Thumbnail
            videos_database = {
                'Video_ID': Video_ID,
                'CHANNEL_ID': CHANNEL_ID,
                'Extraction_Date': Extraction_Date,
                'Title': Title,
                'Published_At': Published_At,
                'Description': Description,
                'Thumbnail_url': Thumbnail_url
            }

            videos_database = informacion_especifica(Video_ID, videos_database)
            df_videos = df_videos.append(videos_database, ignore_index=True)
    return df_videos
```

```python
df_relevancia = obtener_informacion_videos_relevance(API_KEY,CHANNEL_ID)
```

```python
# Obtenemos todos los videos
def obtener_informacion_all_videos_fecha(API_KEY, CHANNEL_ID) -> pd.DataFrame:
    num_pagina = 1
    df_videos = pd.DataFrame(columns=['Video_ID','CHANNEL_ID','Extraction_Date','Title','Published_At','Description','Thumbnail_url','Duration','Likes','Views','Comments_Count','Favorites_Count','Tags'])
    order = 'date'
    pageToken = ''
    while True:
        url = f'https://www.googleapis.com/youtube/v3/search?key={API_KEY}&channelId={CHANNEL_ID}&part=snippet,id&maxResults=50&order={order}&pageToken={pageToken}' # Para buscar videos
        response = requests.get(url).json()
        resultados_por_pagina = response['pageInfo']['resultsPerPage']
        resultados_totales = response['pageInfo']['totalResults']
        print(f'Page {num_pagina} of {int(resultados_totales/resultados_por_pagina)}') # Progreso

        for video in response['items']:
            time.sleep(1)
            if video['id']['kind'] == 'youtube#video':
                Video_ID = video['id']['videoId'] # ID del video
                Extraction_Date = datetime.date.today() # Dia y hora de la extraccion
                Title = video['snippet']['title'] # Titulo del video
                Published_At = video['snippet']['publishedAt'] # Fecha de publicacion
                try:
                    Description = video['snippet']['description'] # Descripcion del video
                except KeyError:
                    Description = np.nan
                Thumbnail_url = video['snippet']['thumbnails']['default']['url'] # El url del Thumbnail
                videos_database = {
                    'Video_ID': Video_ID,
                    'CHANNEL_ID': CHANNEL_ID,
                    'Extraction_Date': Extraction_Date,
                    'Title': Title,
                    'Published_At': Published_At,
                    'Description': Description,
                    'Thumbnail_url': Thumbnail_url
                }

                videos_database = informacion_especifica(Video_ID, videos_database)
                df_videos = df_videos.append(videos_database, ignore_index=True)
        try:
            pageToken = response['nextPageToken']
            num_pagina += 1
        except KeyError:
            print('Se ha finalizado la extraccion de datos')
            break
        
    return df_videos
```

```python
df_all_videos = obtener_informacion_all_videos_fecha(API_KEY,CHANNEL_ID)
```
